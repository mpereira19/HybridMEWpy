```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
```

```python
filename = 'n5000s2.csv'
title = ' sigma=2'
```

```python
def summary(df, format='df'):
    stats = {}
    for cname in df.columns:
        values = df.loc[:,cname].values
        try:
            _means = np.mean(values)
            _vars = np.var(values)
            _maxs = np.max(values)
            _mins = np.min(values)
        except Exception:
            _means = _vars = _maxs = _mins = np.NAN
        stat = {'mean': _means,
                'var': _vars,
                'min': _mins,
                'max': _maxs
                }
        stats[cname] = stat
    if format == 'df':
        import pandas as pd
        df = pd.DataFrame(stats).transpose()
        return df
    else:
        return stats
```

```python
df_s1 = pd.read_csv(filename)
df_s1 = df_s1.drop(columns=['ID'])
df_s1.dropna(axis = 0, how ='any', thresh = None, subset = None, inplace=False)
s=summary(df_s1)
s.sort_values('var',ascending=False)
```

```python
x = df_s1.loc[:, :].values
x = StandardScaler(with_mean=False).fit_transform(x)
pca1 = PCA(n_components=2)
principalComponents1 = pca1.fit_transform(x)
pca1.explained_variance_ratio_
df_s1.loc[:,df_s1.columns[0]].values
```

```python
pc_s1 = pd.DataFrame(data = principalComponents1
             , columns = [ 'PC1', 'PC2'])
pc_s1
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA'+title, fontsize = 20)
ax.scatter(pc_s1.loc[:, 'PC1'],
           pc_s1.loc[:, 'PC2'])
ax.grid()
```
