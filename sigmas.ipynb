# Sigma values script

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
```

```python
# s=1
df_s1 = pd.read_csv('n10000s1.csv')
df_s1 = df_s1.drop(columns=['ID'])
df_s1.dropna()
df_s1['sigma']=1
df_s1
```

```python
df_s1.max()
```

```python
# s=2
df_s2 = pd.read_csv('n5000s2.csv')
df_s2 = df_s2.drop(columns=['ID'])
df_s2.dropna()
df_s2['sigma']=2
df_s2
```

```python
# s=4
df_s4 = pd.read_csv('n1000s4.csv')
df_s4 = df_s4.drop(columns=['ID'])
df_s4.dropna()
df_s4['sigma']=4
df_s4
```

```python
df = pd.concat([df_s1,df_s2,df_s4])
df = df.dropna()
df.reset_index(drop=True, inplace=True)
df
```

```python
colnames = list(df_s1.columns)
maximum = pd.DataFrame([list(df_s1.max()),list(df_s2.max()),list(df_s4.max()),[1,2,3]], columns=colnames)
maximum = maximum.dropna()
maximum.reset_index(drop=True, inplace=True)
maximum
```

```python
maximum['vPK']
```

### PCA

```python
x = df.drop(columns=['sigma']).loc[:, :].values
x = StandardScaler(with_mean=False).fit_transform(x)
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
```

```python
pca.explained_variance_ratio_
```

```python
principalDf = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2'])
```

```python
finalDf = pd.concat([principalDf, df[['sigma']]], axis = 1)
```

```python
finalDf
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('PC1', fontsize = 15)
ax.set_ylabel('PC2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = [4,2,1]
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
	indicesToKeep = finalDf['sigma'] == target
	ax.scatter(finalDf.loc[indicesToKeep, 'PC1'], finalDf.loc[indicesToKeep, 'PC2'], c = color, s = 50, marker = 'o')
ax.legend([r'$\sigma = 4$',r'$\sigma = 2$',r'$\sigma = 1$'])
ax.grid()
```

To mitigate the effects on the PCA of such high values, we applyed a hyperbolic tangent (tanh) to all the values, allowing to retain the scale of lower values, closer to 0, while condensing the higher ones.

```python
tanh = np.tanh(x)
pca1 = PCA(n_components=2)
principalComponents1 = pca1.fit_transform(tanh)
principalDf1 = pd.DataFrame(data=principalComponents1, columns=['PC1', 'PC2'])
finalDf1 = pd.concat([principalDf1, df[['sigma']]], axis=1)
pca1.explained_variance_ratio_
```

```python
fig1 = plt.figure(figsize = (8,8))
ax1 = fig1.add_subplot(1,1,1)
ax1.set_xlabel('PC1', fontsize = 15)
ax1.set_ylabel('PC2', fontsize = 15)
ax1.set_title('2 component PCA', fontsize = 20)
targets1 = [4,2,1]
colors1 = ['r', 'g', 'b']
for target, color in zip(targets1,colors1):
	indicesToKeep1 = finalDf1['sigma'] == target
	ax1.scatter(finalDf1.loc[indicesToKeep1, 'PC1'], finalDf1.loc[indicesToKeep1, 'PC2'], c = color, s = 50, marker = 'o')
ax1.legend([r'$\sigma = 4$',r'$\sigma = 2$',r'$\sigma = 1$'])
ax1.grid()
```

### Correlation and ANOVA

```python
pc1 = dict()
for c in df.columns:
	pc1[c]=finalDf['PC1'].corr(df[c])
pc2 = dict()
for c in df.columns:
	pc2[c]=finalDf['PC2'].corr(df[c])
```

```python
cor = pd.DataFrame([pc1,pc2]).T.rename(columns={0: "PC1", 1: "PC2"})
```

```python
cor
```

Correlations between the original data and each principal component.

```python
cor.loc[abs(cor['PC1']) > 0.9].sort_values(by='PC1', ascending=False)
```

The first principal component is strongly correlated with seven of the original variables. The first principal component increases with increasing 6-Phosphogluconate dehydrogenase, Glucose-6-phosphate dehydrogenase, Ribulose-phosphate epimerase, Transketolase a and b, Transaldolase, and Ribose-phosphate isomerase. This suggests that these seven fluxes vary together. If one increases, then the remaining ones tend to increase as well. This is in agreement with what we can see in Chassagnole's metabolic network, where these seven reactions represent the entire pentose-phosphate pathway in the model.

```python
cor.loc[abs(cor['PC2']) > 0.9].sort_values(by='PC2', ascending=False)
```

The second principal component on the other hand, is strongly positively correlated with six of the original variables. The second principal component increases with increasing Glyceraldehyde-3-phosphate dehydrogenase, Triptophan synthesis, Phosphoglycerate kinase, Phosphoglycerate mutase, Enolase, and Phosphoenolpyruvate carboxylase. This suggests that these six fluxes vary together. If one increases, then the remaining ones tend to increase as well. This is in agreement with what we can see in Chassagnole's metabolic network, where these six reactions represent the second phase of the glycolysis.

Is there a difference between the means of the groups?

```python
df.groupby(['sigma'])[['vPTS']].count()
```

```python
cor_pc1 = ['vPGDH','vG6PDH','vRu5P','vTKA','vTA','vTKB','vR5PI']
df.groupby(['sigma'])[cor_pc1].mean()
```

```python
from scipy.stats import f_oneway
```

```python
anova = []
for v in cor_pc1:
	s1=df.loc[df['sigma'] == 1][v].values
	s2=df.loc[df['sigma'] == 2][v].values
	s4=df.loc[df['sigma'] == 4][v].values
	anv = f_oneway(s1,s2,s4)
	p = anv.pvalue
	F = anv.statistic
	anova.append({'F':F,'p':p})
```

```python
df_anova=pd.DataFrame(anova,index=cor_pc1)
df_anova
```

```python
cor_pc2 = ['vGAPDH','vTRPSYNTH','vPGK','vrpGluMu','vENO','vpepCxylase']
df.groupby(['sigma'])[cor_pc2].mean()
```

```python
anova2 = []
for v in cor_pc2:
	s1=df.loc[df['sigma'] == 1][v].values
	s2=df.loc[df['sigma'] == 2][v].values
	s4=df.loc[df['sigma'] == 4][v].values
	anv1 = f_oneway(s1,s2,s4)
	p = anv1.pvalue
	F = anv1.statistic
	anova2.append({'F':F,'p':p})
```

```python
df_anova2=pd.DataFrame(anova2,index=cor_pc2)
df_anova2
```

Since all p-values are less than 0.05, we can reject the null hypothesis. This implies that we have sufficient proof to say that there exists a difference in the fluxes due to different sigma values.
